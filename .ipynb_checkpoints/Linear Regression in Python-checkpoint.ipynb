## Linear Regression for Predicting Income Composition of Resources depending upon the number of Schooling years

```python
#Scatter plot for Income composition of resources and number of Schooling years
df.plot(kind='scatter', x='Schooling', y='Income composition of resources')
```

```python
#Line plot of schooling and income compostion of resources
sns.lmplot(x ="Schooling", y ="Income composition of resources", data = df, ci = None) 
```

```python
#Making x and y arrays for Linear Regression from the existing dataframe
X = np.array(df['Schooling']).reshape(-1, 1)
Y = np.array(df['Income composition of resources']).reshape(-1, 1)
```

```python
# Splitting train and test dataset
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)  
```

```python
#Performing Linear Regression
from sklearn.linear_model import LinearRegression 
lr = LinearRegression() 
lr.fit(X_train, Y_train) 
```

```python
#Checking R2 for the linear Regression (Max value can be 1.0)
#It shows how much percentage of data fit our model
print(lr.score(X_test, Y_test)) 
```

```python
#the equation of line is: y = b0 + b1 * x
#Checking the predicted value of b0(Intercept) and b1(slope) for the predicted y
b0 = lr.intercept_
b1 = lr.coef_
print("The intercept of predicted line is", b0)
print("The slope of predicted line is", b1)
```

```python
#Using linear Regression to predict income composition of resources for new values of Schoooling years
X1 = [3.2, 5.6, 10.0, 5.5]
print("The New Schooling years are", X1)
X1 = np.array(X1).reshape(-1, 1)
Y1 = lr.predict(X1)
print("The Predicted income compositon of resources are are:\n", Y1)
```
